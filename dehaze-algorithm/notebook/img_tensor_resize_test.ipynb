```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.transforms import ToTensor
import torch

inputs = ToTensor()(Image.open('/mnt/e/DeepLearningCopies/2023/RIDCP/experiments/AOD-I-HAZE/visualization/image_results/999999/0.png').convert('RGB'))[None, ::]
print(inputs.shape)
plt.imshow(inputs.squeeze().permute(1, 2, 0))
```

```python
wsz = 32
_, _, h_old, w_old = inputs.shape
h_pad = (h_old // wsz + 1) * wsz - h_old
w_pad = (w_old // wsz + 1) * wsz - w_old
inputs = torch.cat([inputs, torch.flip(inputs, [2])], 2)[:, :, :h_old + h_pad, :]
inputs = torch.cat([inputs, torch.flip(inputs, [3])], 3)[:, :, :, :w_old + w_pad]
print(inputs.shape)
plt.imshow(inputs.squeeze().permute(1, 2, 0))
```
